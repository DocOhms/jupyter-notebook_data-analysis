```python
import matplotlib.pyplot as plt

# Data
modified_mortar = [16.85, 16.40, 17.21, 16.35, 16.52, 17.04, 16.96, 17.15, 16.59, 16.57]
unmodified_mortar = [16.62, 16.75, 17.37, 17.12, 16.98, 16.87, 17.34, 17.02, 17.08, 17.27]

# Plot histograms
plt.figure(figsize=(12, 6))

# Histogram for Modified Mortar
plt.subplot(1, 2, 1)
plt.hist(modified_mortar, bins=5, color='blue', edgecolor='black')
plt.title('Histogram of Modified Mortar')
plt.xlabel('Tension Bond Strength')
plt.ylabel('Frequency')

# Histogram for Unmodified Mortar
plt.subplot(1, 2, 2)
plt.hist(unmodified_mortar, bins=5, color='green', edgecolor='black')
plt.title('Histogram of Unmodified Mortar')
plt.xlabel('Tension Bond Strength')
plt.ylabel('Frequency')

# Show plots
plt.tight_layout()
plt.show()



# Both data sets in Histogram
import matplotlib.pyplot as plt

# Data
modified_mortar = [16.85, 16.40, 17.21, 16.35, 16.52, 17.04, 16.96, 17.15, 16.59, 16.57]
unmodified_mortar = [16.62, 16.75, 17.37, 17.12, 16.98, 16.87, 17.34, 17.02, 17.08, 17.27]

# Plot histogram
plt.figure(figsize=(10, 6))

# Histogram for Modified Mortar
plt.hist(modified_mortar, bins=5, alpha=0.5, label='Modified Mortar', color='blue', edgecolor='black')

# Histogram for Unmodified Mortar
plt.hist(unmodified_mortar, bins=5, alpha=0.5, label='Unmodified Mortar', color='green', edgecolor='black')

# Adding titles and labels
plt.title('Histogram of Tension Bond Strength')
plt.xlabel('Tension Bond Strength')
plt.ylabel('Frequency')

# Adding a legend
plt.legend(loc='upper left')

# Show plot
plt.tight_layout()
plt.show()

```
