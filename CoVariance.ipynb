```python
import pandas as pd

# Load the dataset
data = pd.read_csv('HistoricalData_1725204029226.csv')

# Date
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)
```

```python
# Calculate the covariance matrix between 'Close/Last' and 'Open'
covariance_matrix = data[['Close/Last', 'Open']].cov()

# Extract the covariance between 'Close/Last' and 'Open'
covariance_value = covariance_matrix.loc['Close/Last', 'Open']

print("Covariance between Close/Last and Open:", covariance_value)
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# sample data
data = pd.read_csv('HistoricalData_1725204029226.csv')
features = data[['Close/Last', 'Open', 'High', 'Low']].dropna()

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3)
data['Cluster'] = kmeans.fit_predict(features)

# Plot clusters
plt.scatter(data['Close/Last'], data['Open'], c=data['Cluster'], cmap='viridis')
plt.xlabel('Close/Last')
plt.ylabel('Open')
plt.title('KMeans Clustering')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# sample data
data = pd.read_csv('HistoricalData_1725204029226.csv')
features = data[['Close/Last', 'Open', 'High', 'Low']].dropna()

# clustering
kmeans = KMeans(n_clusters=3, random_state=42)
data['Cluster'] = kmeans.fit_predict(data[['Close/Last', 'Open', 'High', 'Low']])

# Plotting clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Close/Last', y='Open', hue='Cluster', data=data, palette='viridis', marker='o')
plt.title('Cluster Spread Visualization')
plt.xlabel('Close/Last')
plt.ylabel('Open')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```
