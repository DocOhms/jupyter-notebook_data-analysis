```python
from statistics import mean, stdev, median, variance
import scipy.stats as stats
from matplotlib import pyplot as plt

# Data
nerves = [6625, 6000, 5450, 5200, 5175, 4900, 4750, 4500, 3985, 900, 450, 2800]
muscles = [3900, 3500, 3450, 3200, 2980, 2800, 2500, 2400, 2200, 1200, 1150, 1130]

# Normalization function
def normalize(data):
    min_val = min(data)
    max_val = max(data)
    return [(x - min_val) / (max_val - min_val) for x in data]

# Normalize the data
nerves_normalized = normalize(nerves)
muscles_normalized = normalize(muscles)

# Convert normalized muscle data to percentages of normalized nerve data
def convert_to_percentage(nerves_normalized, muscles_normalized):
    return [(m / n) * 100 if n != 0 else 0 for m, n in zip(muscles_normalized, nerves_normalized)]

muscles_percentage = convert_to_percentage(nerves_normalized, muscles_normalized)

# Scatter Plot normalized data with labels
plt.figure(figsize=(12, 8))
plt.scatter(nerves_normalized, muscles_percentage, color='blue', marker='o')

# Add labels
for i in range(len(nerves)):
    plt.text(nerves_normalized[i], muscles_percentage[i], 
             'Nerve' if i < len(nerves) // 2 else 'Muscle',
             fontsize=9, ha='right')

plt.title('Scatter Plot of Normalized Nerves vs. Percentage of Normalized Values')
plt.xlabel('Normalized Values')
plt.ylabel('Percentage of Normalized Values')
plt.grid(True)
plt.show()


```
