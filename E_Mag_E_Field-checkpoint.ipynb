```python
import matplotlib.pyplot as plt

# Function to plot the line distribution
def plot_line_distribution():
    plt.figure(figsize=(8, 6))

    # Sphere properties
    sphere_radius = 0.5e-3  # radius in meters
    sphere_center = (0, 1)  # center position of the sphere (x, y)

    # Line distribution properties
    line_positions = [-0.05, 0.05]  # x positions of the lines
    line_length = 1.5  # length of lines for visualization

    # Plot the sphere
    circle = plt.Circle(sphere_center, sphere_radius, color='orange', label='Quartz Sphere (+100 mC)')
    plt.gca().add_patch(circle)

    # Plot line distribution
    for pos in line_positions:
        plt.plot([pos, pos], [0, line_length], color='blue', lw=2)
        plt.text(pos, line_length + 0.05, 'Line Charge', ha='center', color='blue')

    # Add annotations
    plt.text(0, 0.85, 'Height: 1 m', fontsize=10, ha='center')
    plt.text(0, 1.15, 'Suspended Height', fontsize=10, ha='center')

    # Set limits and labels
    plt.xlim(-0.15, 0.15)
    plt.ylim(0, 1.5)
    plt.axhline(0, color='k', lw=0.5, ls='--')
    plt.axhline(1, color='k', lw=0.5, ls='--')
    plt.title('Line Distribution Method')
    plt.xlabel('X Position (m)')
    plt.ylabel('Y Position (m)')
    plt.gca().set_aspect('equal', adjustable='box')
    plt.grid()

    # Display calculations
    plt.figtext(0.15, 0.1, "Electric Field (E) = 0.136 N/C\n"
                             "Linear Charge Density ()  9.55  10 C/m\n"
                             "Weight (W) = 0.0136 N\n"
                             "Sphere Charge (q) = +100 mC", 
                fontsize=10, bbox=dict(facecolor='white', alpha=0.5))

    plt.tight_layout()
    plt.show()

# Function to plot the surface distribution
def plot_surface_distribution():
    plt.figure(figsize=(8, 6))

    # Sphere properties
    sphere_radius = 0.5e-3  # radius in meters
    sphere_center = (0, 1)  # center position of the sphere (x, y)

    # Plate distribution properties
    plate_y_positions = [0.9, 1.1]  # y positions of the plates

    # Plot the sphere
    circle = plt.Circle(sphere_center, sphere_radius, color='orange', label='Quartz Sphere (+100 mC)')
    plt.gca().add_patch(circle)

    # Plot surface distribution
    for pos in plate_y_positions:
        plt.plot([-0.1, 0.1], [pos, pos], color='green', lw=5)
        plt.text(0.15, pos, 'Charged Plate', va='center', color='green')

    # Add annotations
    plt.text(0, 0.85, 'Height: 1 m', fontsize=10, ha='center')
    plt.text(0, 1.15, 'Suspended Height', fontsize=10, ha='center')

    # Set limits and labels
    plt.xlim(-0.15, 0.15)
    plt.ylim(0, 1.5)
    plt.axhline(0, color='k', lw=0.5, ls='--')
    plt.axhline(1, color='k', lw=0.5, ls='--')
    plt.title('Surface Distribution Method')
    plt.xlabel('X Position (m)')
    plt.ylabel('Y Position (m)')
    plt.gca().set_aspect('equal', adjustable='box')
    plt.grid()

    # Display calculations
    plt.figtext(0.15, 0.1, "Electric Field (E) = 0.136 N/C\n"
                             "Surface Charge Density ()  2.40  10 C/m\n"
                             "Weight (W) = 0.0136 N\n"
                             "Sphere Charge (q) = +100 mC", 
                fontsize=10, bbox=dict(facecolor='white', alpha=0.5))

    plt.tight_layout()
    plt.show()

# Execute plotting functions
plot_line_distribution()
plot_surface_distribution()

```
