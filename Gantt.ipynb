```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime

# Define tasks and their start/end dates
tasks = {
    'Optimize Experimental Parameters': (datetime(2024, 11, 25), datetime(2024, 12, 14)),
    'Nanowire Deposition and Testing': (datetime(2024, 12, 15), datetime(2024, 12, 31)),
    'Device Integration (Flipper Zero)': (datetime(2025, 1, 1), datetime(2025, 1, 14)),
    'Noise Reduction and Sensitivity Testing': (datetime(2025, 1, 15), datetime(2025, 1, 31)),
    'Prototype Evaluation and Improvements': (datetime(2025, 2, 1), datetime(2025, 2, 28)),
    'Final Product Development': (datetime(2025, 3, 1), datetime(2025, 3, 31)),
    'Documentation and Reporting': (datetime(2025, 3, 15), datetime(2025, 4, 15))
}

# Create the plot
fig, ax = plt.subplots(figsize=(12, 6))

# Plotting each task
for task, (start, end) in tasks.items():
    ax.barh(task, end - start, left=start, height=0.5, align='center')

# Formatting
ax.xaxis_date()
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
fig.autofmt_xdate()

# Labels and title
plt.xlabel('Date')
plt.ylabel('Tasks')
plt.title('Projected Timeline')

# Adjust layout to prevent clipping of tick-labels
plt.tight_layout()

plt.show()
```
