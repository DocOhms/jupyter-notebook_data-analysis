```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(-5, 5, 400)
R_LC_ratio = 1
f_t = np.exp(-R_LC_ratio * t)

plt.plot(t, f_t, label=r'$e^{- \frac{R}{LC} t}$')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlabel("Time (t)")
plt.ylabel("f(t)")
plt.legend()
plt.grid()
plt.show()

```
