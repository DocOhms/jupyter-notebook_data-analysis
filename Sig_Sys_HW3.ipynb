```python
import numpy as np
import matplotlib.pyplot as plt

# Define Variables
RC = 2
time = np.linspace(0, 5, 100)  # Time from 0 to 5 seconds
omega = np.linspace(-10, 10, 1000)  # Frequency range from -10 to 10

# Step 1: Calculate x(t)
x_t = -np.exp(-time / RC)

# Step 2: Magnitude and Phase
magnitude = 1 / np.sqrt((RC**2 / 4) + omega**2)
phase = -np.pi - np.arctan(2 * omega / RC)

# phase from radians to degrees
phase_degrees = np.degrees(phase)

# x(t)
plt.figure(figsize=(12, 6))

plt.subplot(3, 1, 1)
plt.plot(time, x_t, label='x(t) = -e^{-t/2}')
plt.title('Response x(t) of the RC Circuit')
plt.xlabel('Time (s)')
plt.ylabel('x(t)')
plt.grid()
plt.legend()

# Plotting Magnitude
plt.subplot(3, 1, 2)
plt.plot(omega, magnitude, label='Magnitude', color='blue')
plt.title('Magnitude of H(jω)')
plt.xlabel('Frequency (ω)')
plt.ylabel('|H(jω)|')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')  
plt.grid()
plt.legend()

# Plotting Phase
plt.subplot(3, 1, 3)
plt.plot(omega, phase_degrees, label='Phase', color='orange')
plt.title('Phase of H(jω) in Degrees')
plt.xlabel('Frequency (ω)')
plt.ylabel('Phase (degrees)')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')  
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define the variables
RC = 2  # Resistance * Capacitance
t = np.linspace(0, 5, 100)  # Time from 0 to 5 seconds
omega = np.linspace(-10, 10, 1000)  # Angular frequency range from -10 to 10 rad/s

# Step 2: Define x(t)
x_t = -np.exp(-t / RC)

# Step 3: Calculate X(s)
# For x(t) = -e^(-t/RC), the Laplace transform X(s) is:
# X(s) = 1 / (s + 1/RC)
s = 1 / RC
X_s = lambda s: 1 / (s + 1/RC)

# Step 4: Set up for frequency response H(jω)
H_jw = lambda omega: 1 / (1j * omega + 1/RC)

# Step 5: Calculate magnitude and phase
magnitude = np.abs(H_jw(omega))  # Magnitude |H(jω)|
phase = np.angle(H_jw(omega))     # Phase in radians
phase_degrees = np.degrees(phase)  # Convert to degrees

# Convert angular frequency (omega) to frequency in Hz
frequencies_hz = omega / (2 * np.pi)

# Step 6: Plot x(t)
plt.figure(figsize=(12, 6))

plt.subplot(3, 1, 1)
plt.plot(t, x_t, label='x(t) = -e^{-t/2}')
plt.title('Response x(t) of the RC Circuit')
plt.xlabel('Time (s)')
plt.ylabel('x(t)')
plt.grid()
plt.legend()

# Step 7: Plot Magnitude
plt.subplot(3, 1, 2)
plt.plot(frequencies_hz, magnitude, label='Magnitude', color='blue')
plt.title('Magnitude of H(jω)')
plt.xlabel('Frequency (f in Hz)')
plt.ylabel('|H(jω)|')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid()
plt.legend()

# Step 8: Plot Phase
plt.subplot(3, 1, 3)
plt.plot(frequencies_hz, phase_degrees, label='Phase', color='orange')
plt.title('Phase of H(jω) in Degrees')
plt.xlabel('Frequency (f in Hz)')
plt.ylabel('Phase (degrees)')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```
